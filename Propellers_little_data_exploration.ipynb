```python
# 1 Вариант
data_filtered = data.loc[data["conversions"] <= data["impressions"]]
```

```python
data_filtered["CTR"] = data_filtered["conversions"] / data_filtered["impressions"]
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import pearsonr

def cramers_v(var1, var2):
    crosstab =np.array(pd.crosstab(var1, var2, rownames=None, colnames=None)) # Cross table building
    stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
    obs = np.sum(crosstab) # Number of observations
    mini = min(crosstab.shape) - 1 # Take the minimum value between the columns and the rows of the cross table
    return np.sqrt(stat / (obs * mini))

def make_categorical_corr_stat(data, categorical_cols, target_col="CTR"):
    row_list = []
    for colname in categorical_cols:
        contingency_table = pd.crosstab(data[colname], data[target_col])
        chi2, p_value, _, _ = chi2_contingency(contingency_table)
        cramers_v_value = cramers_v(data[colname], data[target_col])
        row_list.append({"Feature Name": colname, "Chi-2": chi2, "P-value": p_value, "Cramers V": cramers_v_value})
    corr_stat = pd.DataFrame(row_list)
    return corr_stat

def make_numerical_corr_stat(data, num_cols, target_col="CTR"):
    row_list = []
    for colname in num_cols:
        corr_coefficient, p_value = pearsonr(data[colname], data[target_col])
        row_list.append({"Feature Name": colname, "Pearsons": corr_coefficient, "P-value": p_value})
    corr_stat = pd.DataFrame(row_list)
    return corr_stat
```

```python
corr_stat1 = make_categorical_corr_stat(data_filtered, categorical_cols, target_col="CTR")
corr_stat1
```

```python
corr_stat2 = make_numerical_corr_stat(data_filtered, num_cols, target_col="CTR")
corr_stat2
```

```python
# почистим дополнительно данные
MIN_IMPRESSIONS = 5
bad_banner_id = stat_block_list[0].loc[stat_block_list[0]["impressions"] < MIN_IMPRESSIONS].index
bad_advertiser_id = stat_block_list[1].loc[stat_block_list[1]["impressions"] < MIN_IMPRESSIONS].index
bad_offer_id = stat_block_list[2].loc[stat_block_list[2]["impressions"] < MIN_IMPRESSIONS].index
bad_geo = stat_block_list[5].loc[stat_block_list[5]["impressions"] < MIN_IMPRESSIONS].index
bad_region_code = stat_block_list[6].loc[stat_block_list[6]["impressions"] < MIN_IMPRESSIONS].index
bad_city_code = stat_block_list[7].loc[stat_block_list[7]["impressions"] < MIN_IMPRESSIONS].index
bad_lang = stat_block_list[14].loc[stat_block_list[14]["impressions"] < MIN_IMPRESSIONS].index
bad_isp = stat_block_list[15].loc[stat_block_list[15]["impressions"] < MIN_IMPRESSIONS].index
bad_mobile_isp = stat_block_list[16].loc[stat_block_list[16]["impressions"] < MIN_IMPRESSIONS].index

def cut_data(data):
    cut_condition = ((~data["banner_id"].isin(bad_banner_id)) &
                                      (~data["advertiser_id"].isin(bad_advertiser_id)) &
                                      (~data["offer_id"].isin(bad_offer_id)) &
                                      (~data["geo"].isin(bad_geo)) &
                                      (~data["region_code"].isin(bad_region_code)) &
                                      (~data["city_code"].isin(bad_city_code)) &
                                      (~data["lang"].isin(bad_lang)) &
                                      (~data["isp"].isin(bad_isp)) &
                                      (~data["mobile_isp"].isin(bad_mobile_isp)) &
                                      (~data["geo"].isin(["A1", "A2"])))
    data_filtered = data.loc[cut_condition]
    return data_filtered
```

```python
data_filtered_with_ids = cut_data(data_filtered)
```

В целом, обилие id в признаках это не самая лучшая и интерпертируемая практика,
 можно кодировать категориальные признаки при помощи какого-нибудь подхода, чтобы перевести их в числовые признаки.  
Особенно это неприятно, когда очень много значений в категориальном признаке,
 обычно тогда получается дисбаланс по некоторым, могут быть значения признаков, для которых недостаточно собранно данных по целевой переменной. Тем более если это id, которые каждый день могу появлятюся новые. Лучшим подходом было бы их закодировать какими-нибудь более постоянными храктеристиками характеристиками, например категориями интересов, так будет обеспечена большая стабильность модели.

По данным видно, что довольно много строк содержат высокие конверсии, потому что был совершен клик при очень малом кол-ве показов, а для некоторых значений признаков и вовсе 1 клик при 1 показе. Обученная на таких данных модель может быть очень нестабильна.

По корреляциям можно сказать, что могут быть важными признакими *zone_id*, *affiliate_id*, *browser*, *hour* по значениям Хи-квадарат (но из-за большой степени свободы корреляция может быть слабой), а по V Крамера - у признаков довольна слабая корреляция, но наибольшей обладают *city_code*, *banner_id*, *isp* и *advertiser_id*
